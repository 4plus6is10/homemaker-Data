```python
import os
import pandas as pd
import glob
```

```python
files_joined = os.path.join('E://Amazon review data (2018)/100. amazon review data/Furniture/','*.csv')

# Return a list of all joined files
list_files = glob.glob(files_joined)

print("** Merging multiple csv files into a single pandas ratings **")
# Merge files by joining all files
ratings2 = pd.concat(map(pd.read_csv, list_files), ignore_index=True)
ratings2
```

```python
ratings2 = ratings2[['ID', 'Asin', 'Rating']]

ratings2
```

```python
asinList = list(set(ratings2['Asin']))
asinList
```

```python
asinNumber = {}
idx = 0

for i in asinList:
    asinNumber[i] = idx
    idx += 1
        
asinNumber
```

```python
asinNumber.get('B09KGZ1SJF')
```

```python
numberList = []
for i in ratings2['Asin']:
    numberList.append(asinNumber.get(i))
    
numberList
```

```python
ratings2['AsinNumber'] = numberList
ratings2
```

```python
ratings2 = ratings2[['ID', 'AsinNumber', 'Rating']]
ratings2
```

```python
ratings2.to_csv('./ratings2Furniture.csv', sep=',', index=False, header=False)
```

```python
from surprise import SVD
from surprise import Dataset
from surprise import accuracy
from surprise import Reader
from surprise.model_selection import train_test_split
```

```python
from surprise import Reader
 
reader = Reader(line_format='user item rating', sep=',', rating_scale=(1, 5))
data=Dataset.load_from_file('ratings2.csv',reader=reader)
```

```python
trainset, testset = train_test_split(data, test_size=.25, random_state=0)

# 수행시마다 동일한 결과 도출을 위해 random_state 설정
# algo = SVD(n_factors=50, random_state=0)
algo = SVD()
# 학습 데이터 세트로 학습 후 테스트 데이터 세트로 평점 예측 후 RMSE 평가
algo.fit(trainset)
predictions = algo.test( testset )
accuracy.rmse(predictions)
```
